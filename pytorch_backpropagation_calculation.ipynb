```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
input = torch.tensor( [ [0.1 , 0.2 , 0.7 ] ])
```

```python
bias = torch.tensor( [ [1.0 , 1.0 , 1.0 ] ])
```

```python
w1 = torch.tensor(  [ 
                        [0.1 , 0.2 , 0.3],
                        [0.3 , 0.2 , 0.7],
                        [0.4 , 0.3 , 0.9]
                        ])
```

```python
w2 = torch.tensor(  [ 
                        [0.2 , 0.3 , 0.5],
                        [0.3 , 0.5 , 0.7],
                        [0.6 , 0.4 , 0.8]
                        ])
```

```python
w3 = torch.tensor(  [ 
                        [0.1 , 0.4 , 0.8],
                        [0.3 , 0.7 , 0.2],
                        [0.5 , 0.2 , 0.9]
                        ])
```

```python
relu = F.relu(input.matmul(w1).add(bias))
sigmoid = torch.sigmoid(relu.matmul(w2).add(bias))
softmax = torch.softmax(sigmoid.matmul(w3).add(bias) , dim=1)
print(softmax)
# [0.19858, 0.28559, 0.51583] 

```
